# Вариант 3. Таблица с данными о товарах


## Импортируем pandas и загружаем нашу таблицу

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/goods_table.csv')
```

## Изучаем данные и их структуру - смотрим первые 10 позиций

```python
df.head(10)
```

## Добавляем столбец "Цена со скидкой" и сортируем по нему в порядке убывания

```python
df['Цена со скидкой'] = df['Цена'] * (1 - df['Скидка в процентах'] / 100)
df = df.sort_values(by='Цена со скидкой', ascending=False)
df.head(10)
```

## Группируем по категории товара и вычисляем среднее значение как для изначальной цены, так и для цены со скидкой

```python
grouped_data = df.groupby('Категория')
grouped_data[['Цена', 'Цена со скидкой']].mean()
```
